# ### Week 3 – Day 1: Python Basics  
**Goal:** Understand Python variables, data types, lists, dictionaries, and control flow.  
This is the foundation for building real data-engineering logic in later weeks.


```python
# Basic 1 — Variables & Data Types
x = 10
y = 3.5
z = "Data Engineer"

print(type(x), type(y), type(z))
```

# Explanation  
- `type()` shows the variable’s data type.  
- Strings are text; integers and floats are numbers.  
- Understanding these basics is key before manipulating data in Pandas later.


```python
# Basic 2 — Lists & Dictionaries
items = ["apple", "banana", "mango"]
prices = {"apple": 1.2, "banana": 0.8, "mango": 1.5}

print(items[0])
for fruit, price in prices.items():
    print(f"{fruit} -> ${price}")
```

# Explanation  
- Lists store ordered values and are indexed starting at 0.  
- Dictionaries hold **key–value pairs** (like column–value mapping).  
- You’ll use them constantly in ETL config files and JSON parsing.


```python
# Basic 3 — Control Flow
for fruit, price in prices.items():
    if price > 1:
        print(f"{fruit} is premium")
    else:
        print(f"{fruit} is cheap")
```

# Explanation  
`if/else` statements let your program make decisions.  
They’re the foundation for filtering and validation logic in pipelines.


```python
# Basic 4 — Challenge: Average Comparison
numbers = [12, 5, 20, 3, 17]
avg = sum(numbers)/len(numbers)

for n in numbers:
    print(f"{n} is {'above' if n > avg else 'below'} average")
```

# Challenge Explanation  
- `sum()` and `len()` compute totals and counts.  
- The inline `if` (`'above' if n > avg else 'below'`) is a compact way to apply logic.  
- This pattern is common in data-quality or rule-based checks.
